```python
df2['Amount']= round(df2['Amount'],2)
df2
```

```python
df2.groupby(['Type'], sort=False)['Amount'].mean().sort_values(ascending=False)

# Heighest category spend up on is Car
```

```python
df_dumm=df2.groupby(['Customer'],as_index=False)['Amount'].sum()
```

```python
df_dumm
```

```python
pd.merge(df1,df_dumm,on='Customer',  how='left')
```

```python
df3_merged = pd.merge(df1, df_dumm)
```

```python
 df3_merged['percnt_Spend']= round((df3_merged.Amount/df3_merged.Limit)*100,2)
```

```python
df3_merged
```

```python
df3_merged[df3_merged['percnt_Spend'] > 90]

#most customers spend more than 90% of their spending
```

```python
bins = [18, 30, 40, 50, 60, 70, 120]
labels = ['18-29', '30-39', '40-49', '50-59', '60-69', '70+']
df3_merged['agerange'] = pd.cut(df3_merged.Age, bins, labels = labels,include_lowest = True)
```

```python
df3_merged.drop('Age', axis=1, inplace=True)
```

```python
df3_merged
```

```python
df3_merged.groupby(['agerange'], sort=False)['Amount'].sum().sort_values(ascending=False)

# Age group 18-29 spends the most
```
